```python
import random

print("=== Escape the Vault🔒 ===")
print("You are trapped inside a vault.")
print("Guess the correct code to escape!")

vault_code = random.randint(1, 10) # random number generator betwen 1-10
attempts = 3 # player has 3 attempts to guess 

while attempts > 0:# loops continues until three attempts are over
    try:
        player_guess = int(input("Enter the vault code (1-10): "))

        if player_guess == vault_code: #checks if player guessed correcttly
            print("Vault opened! You escaped!")
            break

        elif player_guess < vault_code:
            print("Too low!")

        else:
            print("Too high!")

        attempts -= 1 # this will reduce the number of attempts when the guess is wrong 
        print("Attempts left:", attempts)
        
    except ValueError:
        print("Invalid input! ENTER A NUMBER!")# Runs if player enters something that isn' a number

if attempts == 0:#if the player runs out of attempts then the game ends
    print("🚨 ALARM! Game over.")
    print("The correct code was:", vault_code)
```
