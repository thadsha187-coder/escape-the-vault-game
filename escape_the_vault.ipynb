```python
import random

#tracks the score throughout game
total_games = 0
total_wins = 0
total_losses = 0

def choose_difficulty():
    #ask the player to choose which level they want to play
    difficulty=input("Choose difficulty (easy or medium or hard):").lower()
   
    if difficulty =="easy":
        max_number =10
        attempts = 5
       
    elif difficulty =="medium":
        max_number = 30
        attempts= 4
   
    elif difficulty =="hard":
        max_number = 50
        attempts= 3
    else:
        print("Invalid choice.Defaulting to easy.")
        max_number =10
        attempts=5
    return max_number, attempts

def show_stats():
    # display the player's overall game statistics
    print("\n=== GAME SCORE ===")
    print("Games Played:", total_games)
    print("Wins:", total_wins)
    print("Losses:", total_losses)




while True:

    print("\n=== Escape the Vault 🔒 ===")
    print("You wake up trapped inside a high-security vault.")
    print("The only way out is to crack the secret code.")
    print("Choose your difficulty and escape before the alarm triggers!\n")

    # Start a new round
    max_number, attempts = choose_difficulty()
    vault_code = random.randint(1, max_number)

    total_games += 1
    print(f"\nNew round started! (Range: 1-{max_number})")
    print(f"You have {attempts} attempts.\n")

    while attempts > 0:# loops continues until attempts are over
        try:
            player_guess = int(input(f"Enter the vault code (1-{max_number} ): "))

            if player_guess < 1 or player_guess > max_number:
                print(f"Please enter a number between 1 and {max_number}.")
                continue  # goes back to the top of the loop without losing an attempt
           
            elif player_guess == vault_code: #checks if player guessed correcttly
                print("Vault opened! You escaped!")
                total_wins += 1
                break

            elif player_guess < vault_code:
                print("Too low!")

            else:
                print("Too high!")

            attempts -= 1 # this will reduce the number of attempts when the guess is wrong
            print("Attempts left:", attempts)
       
        except ValueError:
            print("Invalid input! ENTER A NUMBER!")# Runs if player enters something that isn' a number

    if attempts == 0:#if the player runs out of attempts then the game ends
            print("🚨 ALARM! Game over.")
            print("The sirens go off and the vault seals shut.")
            print("The correct code was:", vault_code)
            total_losses += 1
   
 #ask if they want to play again
    play_again = input('Play again?(yes/no):').lower()
    if play_again != "yes":
        show_stats()
        print("Thanks for playing Escape the Vault 🔒")
        break
        
```
