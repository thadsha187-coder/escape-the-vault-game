```python
import random

print("=== Escape the Vault ===")
print("You are trapped in a vault.")
print("Guess the correct code to escape!")

vault_code = random.randint(1,10)

player_guess = int(input("Enter the vault code (1-10): "))

if player_guess == vault_code:
    print("Vault opened! You escaped!")
else:
    print("Wrong code! The vault remains locked.")
```
